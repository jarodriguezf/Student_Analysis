```python
# Creamos el grafico con las distribuciones de cada caracteristica
fig, axes = plt.subplots(5, 7, figsize=(50, 35))
axes = axes.flatten() # Aplanar el arreglo de subgráficos

for i, feature in enumerate(df.columns):
    # Contar las ocurrencias de cada valor en la característica
    value_counts = df[feature].value_counts()

    # Obtenemos la columna y el valor de cada caracteristica
    values_to_plot = value_counts.index
    frequencies = value_counts.values

    # Crear bar plot en el subgráfico correspondiente
    axes[i].bar(values_to_plot, frequencies)
    axes[i].set_title(feature,  fontsize=30)
    axes[i].tick_params(axis='both', which='major', labelsize=30)

# Ajustar y mostrar graficos
plt.tight_layout()
plt.show()
```

**Correlacion y graficos multivariable**

```python
# Creamos un mapa de calor con las correlaciones mas significativas (los valores en torno a 0 se ocultan)
numeric_values = df.select_dtypes(include='number').corr()
trimask = np.triu(np.ones_like(numeric_values, dtype=bool))
plt.figure(figsize=(10, 10))
sns.heatmap(data=numeric_values,cmap='RdYlGn_r',mask=trimask | ((np.abs(numeric_values) <= 0.3) & (np.abs(numeric_values) >= -0.2)), 
                                                        annot=False, linewidths=.5, cbar_kws={'shrink': 0.5})
plt.show()
```

```python
# Graficamos variables correlacionadas con mayores coeficientes.
variable_values = ["Age at enrollment","Daytime/evening attendance","International","Nacionality","Father's occupation",
                   "Mother's occupation","Displaced","Tuition fees up to date","Debtor","Curricular units 1st sem (enrolled)",
                   "Curricular units 1st sem (credited)","Curricular units 2nd sem (credited)","Curricular units 2nd sem (enrolled)",
                   "Curricular units 2nd sem (approved)","Curricular units 2nd sem (grade)",'Curricular units 1st sem (approved)',
                   "Curricular units 1st sem (grade)","Curricular units 1st sem (evaluations)","Curricular units 2nd sem (evaluations)",
                   "GDP","Unemployment rate"]

# Creamos un pairplot utilizando sns.pairplot() con las características filtradas
sns.pairplot(df[variable_values], diag_kind="kde", corner=True)
```

**Analisis multivariables**

*Procederé a realizar un analisis de las relaciones fuertes tanto positivas como negativas entre las caracteristicas vistas:*
- Age at enrollment vs Daytime/evening attendance: Relación negativa.
- International vs Nacionality: Relación positiva fuerte (si el estudiante es internacional, se supone que su nacionalidad no es local).
- Fathers occupation vs Mothers occupation: Relación positiva (La ocupacion del padre del estudiante tiende a ser positiva con respecto a la madre, por tanto, ambos padres se situan en puestos similares).
- Age at enrollmentr vs Displaced: Relación negativa.
- Tuition fees up to date vs Debtor: Relacion negativa.
- Curricular units 1st sem (enrolled) vs Curricular units 1st sem (credited): Relación positiva fuerte (la mayoria a aprobado las asignaturas que cursaban).
- Curricular units 1st sem (credited) vs Curricular units 2nd sem (credited): Relacion fuerte positiva (la mayoria tiende a completar las mismas asignaturas en ambos semestres).
- Curricular units 1st sem (enrolled) vs Curricular units 2nd sem (enrolled): Relación fuerte positiva (la mayoria tiende a elegir la misma cantidad de asignaturas en ambos semestres).
- Curricular units 1st sem (approved) vs Curricular units 2nd sem (approved): Relación fuerte positiva (la mayoria aprueban ambos semestres de la misma manera).
- Curricular units 1st sem (grade) vs Curricular units 2nd sem (grade): Relacion positiva fuerte (la mayoria tienden a obtener las mismas calificaciones en ambos semestres).
- GDP vs Unemployment rate: Relacion negativa.
