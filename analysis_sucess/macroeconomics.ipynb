#### **Análisis macroeconomico exito academico**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.display.max_columns = None
```

```python
df = pd.read_csv('../data_raw/dataset.csv')
df
```

**Extraer características**

```python
macroeconomics_features = df.loc[df['Target']=='Graduate',['Unemployment rate','Inflation rate','GDP']]
```

```python
macroeconomics_features.describe()
```

```python
for i in macroeconomics_features.columns:
    sns.histplot(data=macroeconomics_features, x=i, kde=True)
    plt.show()
```

Observaciones:
- Los estudiantes vivieron una tasa de desempleo del 11.
- La inflacion se situaba entre 0.5-1.5.
- El PIB se situaba en el 1.7.

```python
corr_macro = macroeconomics_features.corr(method='spearman')
sns.heatmap(data=corr_macro,cmap='RdYlGn_r', annot=True, linewidths=.5)
plt.title('Datos Macroeconómicos')
plt.show()
```

```python
sns.pairplot(macroeconomics_features, diag_kind="kde")
```

- Al aumentar la tasa de desempleados tiénde a disminuir el GDP (PIB), esto el logico ya que con menos mano de obra, el país produce menos.
- Cuando la inflación del pais es baja, el desempleo tiende a aumentar y viceversa.
- Con respecto al GPD y la inflación, podemos observar como un poco de inflación ayuda a aumentar el GDP, si la inflación excede la tasa de 1, el GDP vuelve a caer.

### Conclusiones
El estudio y análisis de los estudiantes en el marco macroeconómico dado es el siguiente:
- Los estudiantes que se graduaron de sus estudios, vivieron una epoca de desempleo moderado, pero a causa de la inflación seguía creciendo el desempleo.
- La inflación ya estaba suponiendo un problema para las economías.
- El PIB dada la inflación superior a 1, cayó.
